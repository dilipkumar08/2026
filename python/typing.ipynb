```python
from typing import List, Dict, Tuple, Set, Any, Optional,Iterable


#when we want to specify the type of data we are passing to a function we can use the above datatypes

def print_data(data: List[int]|Tuple|Set|Dict) -> Any:
    print(*data)
    return len(data)

# so instead of writing multiple datatypes for the iteration we can use Iterable
# which is a generic type that can accept any iterable 
# data type like list, tuple, set, dict etc.

def process_data(data:Iterable) -> Optional[int]:
    for item in data:
        print(item)
    return len(data)

if __name__=="__main__":
    print_data([1, 2, 3, 4, 5])
    process_data({1, 2, 3, 4, 5})
```

```python
from typing import Protocol
from abc import ABC, abstractmethod 


class Animal(ABC):
    @property
    @abstractmethod
    def legs(self)->int:
        pass
    
    @abstractmethod
    def make_sound(self)->str:
        pass

# so why do we use protocol??
# we use protocol when we want to define a common interface for 
# different classes without using inheritance.
# for example we can have a class snake that makes a sound but it doesn't have legs 
# so it can't inherit from the Animal class but we can define a protocol for 
# it and make it implement the make_sound method.

class Sound(Protocol):
    def make_sound(self) -> str:
        pass

class snake:
    def make_sound(self) -> str:
        return "Hiss"
        
    
class Dog(Animal):
    def make_sound(self) -> str:
        return "Bark"
    @property
    def legs(self) ->int:
        return 4
   
def getanimal_sound(animal: Sound) -> None:
    for sound in animal:
        print(sound.make_sound())
    
if __name__=="__main__":
    dog=Dog()
    snake=snake()
    getanimal_sound([dog, snake])
    
```

```python
from typing import Protocol, TypeVar , List, Iterable
from abc import ABC, abstractmethod

T=TypeVar('T', covariant=True)

class Counter(ABC):
    @abstractmethod
    def count(self,str) -> int:
        pass
    
    @abstractmethod
    def count_int(self,data: int) -> int:
        pass
    

class Counter1(Counter):
    def count(self,data: str) -> int:
        return len(data)
    
    def count_int(self,data: int) -> int:
        return self.count(str(data))


class CounterProtocol(Protocol[T]):
    def count(self,str) -> T:
        pass

class Counter2():
    def count(self,data: List)-> int:
        return len(data)
    


def get_count(counter: Iterable[CounterProtocol]) -> int:
    for Counter in counter:
        print(Counter.count("212"))




if __name__=="__main__":
    get_count([Counter1(), Counter2()])
```

```python
from typing import Protocol, TypeVar, Any

T=TypeVar('T', covariant=True)

class MySequence(Protocol[T]):
    def __getitem__(self, __idx: Any) -> T:
        ...
    
    def __len__(self) -> int:
        ...

def process_sequence(seq: MySequence[str]) -> str:
    if len(seq) > 0:
        return seq[0]
    return "Empty sequence"
```

```python
from typing import Protocol, TypeVar, Any
from  typing_extensions import Self

class Lessthan(Protocol):
    def __lt__(self,other:Self)->bool:
        pass

T=TypeVar('T',bound=Lessthan)



def find_minimum(a:T,b:T)->bool:
    if a < b:
        return a
    return b


if __name__=="__main__":
    print(find_minimum(3, 5))
    print(find_minimum("apple", "banana"))
```

```python
from dataclasses import dataclass
from typing import Self

@dataclass
class Line:
    xmin:float
    xmax:float
    def __lt__(self,other:Self)->bool:
        return (self.xmax - self.xmin) < (other.xmax - other.xmin)

if __name__=="__main__":
    line1=Line(0, 12)
    line2=Line(0, 10)
    print(max(line1,line2))
```

```python
#collections
from typing import List, Dict, Tuple, Set, Optional

def print_data(data: Dict[str|int,float|int],data2: List[Tuple[str,int]|Tuple[int,int]],data3:Set[float]|None=None)->None:
    for key,value in data.items():
        print(f"{key}: {value}")
    
    for item in data2:
        print(*item)

    if data3 is not None:
        print(*data3)

if __name__=="__main__":
    print_data({"a": 1, "b": 2}, [("x", 10), (20, 30)])

    print_data({"a": 1, "b": 2}, [("x", 10), (20, 30)], {3.14, 2.71})
```

```python
#Containers
from typing import Iterable
from typing import Iterator
from typing import Sequence
from typing import Any

def get_index(data:Sequence,index:int)->Any:
    return data[index]


def iter_data(data:Iterator)->None:
        try:
            while True:
                print(next(data))
        except StopIteration:
            print("End of iteration")

def loop_data(data:Iterable)->None:
    for item in data:
        print(item)

if __name__=="__main__":
    loop_data([1, 2, 3, 4, 5])
    loop_data((1, 2, 3, 4, 5))
    loop_data({1, 2, 3, 4, 5})
    loop_data({"a": 1, "b": 2})
    iter_data(iter([1, 2, 3, 4, 5]))
    print(get_index([1, 2, 3, 4, 5], 2))





```

Overload is for type checking for example 
if a two functions with same number of params but just the dtype is different
we can use overload to differentiate them just like overloading concept.

```python
from typing import overload

class Concat:
    @overload
    def concat(self,a:str,b:str)->str:
        pass
    
    @overload
    def concat(self,a:int,b:int)->int:
        pass
    
    @overload
    def concat(self,a:List[int],b:List[int])->List[int]:
        pass

    def concat(self,a: int|str|List[int],b:int|str|List[int])->int|str|List[int]:
        return a+b
    
if __name__=="__main__":
    concat=Concat()
    print(concat.concat("Hello ", "World"))
    print(concat.concat(10, 20))
    print(concat.concat([1, 2], [3, 4]))
```

```python
from sklearn.ensemble import AdaBoostClassifier,RandomForestClassifier,RandomForestRegressor
from sklearn.model_selection import train_test_split

from typing import Protocol
import numpy as np



class Model(Protocol):
    def fit(self,x_train:np.array,y_train:np.array)->None:
        ...
    def predict(self,x_test:np.array)->np.array:
        ...


def evaluate(model:Model,x_test:np.array,y_test:np.array):
    preds=model.predict(x_test)
    return preds

if __name__=="__main__":
    model1=AdaBoostClassifier()
    model2=RandomForestClassifier() 
    model3=RandomForestRegressor()

    x=np.array([1,2,3,4,5,6,7,8,10])
    y=np.array([1,2,3,4,5,6,7,8,10])
    x=x.reshape(-1, 1)
    y=y.reshape(-1)

    print(x,y)

    x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,shuffle=False)

    print(x_train,x_test)
    print(y_train,y_test)

    models=[model1,model2,model3]

    for model in models:
        model.fit(x_train,y_train)
        print(f"model:{model},{y_test} ",evaluate(model,x_test,y_test))




```

so, Final is to define a constant value
and final is to define a constant function which cannot be modified by child class

```python
from typing import final, Final

class User:
    company_name: Final="ABC company"
    def __init__(self,name:str,role:str):
        self.name=name
        self.role=role

    @final
    def user_desc(self):
        print(f"{self.name} is a {self.role} at {self.company_name}")

if __name__=="__main__":
    User("Dilip","Software developer").user_desc()
```
