#### Classes and Objects

```python
class Animal:
    def __init__(self,name):
        self.name=name
    
    def sound(self):
        if self.name=="dog":
            return "woof"
        elif self.name=="cat":
            return "meow"
        else:
            return "unknown sound"

if __name__=="__main__":
    dog=Animal("dog")
    cat=Animal("cat")
    cow=Animal("cow")

    print(f"{dog.name} says {dog.sound()}")
    print(f"{cat.name} says {cat.sound()}")
    print(f"{cow.name} says {cow.sound()}")
```

so, we can make use the class to group similar functions under a category and call them using it's instance which is called objects.

#### Inheritance

```python
class Vehicle:
    def __init__(self,type: str,tire:str):
        self.type=type
        self.tire=tire
    
    def time_to_travel(self,distance:int,speed:int):
        if speed<=0:
            return "Speed must be greater than zero"
        time=distance/speed
        return f"Time to travel {distance} km at {speed} km/h is {time:.2f} hours"
    

class Car(Vehicle):
    def __init__(self,model,engine:str,year:int):
        super().__init__(type="car",tire="4")
        self.model=model
        self.engine=engine
        self.year=year
    
    def car_info(self):
        return f"{self.year} {self.model} with {self.engine} engine"
    
    def time_to_travel(self, distance:int, speed:int):
        return super().time_to_travel(distance, speed)

if __name__=="__main__":
    myCar=Car(model="Jaguar",engine="x900",year=2025)
    print(myCar.model)
    print(myCar.engine)
    print(myCar.year)
    print(myCar.type)
    print(myCar.tire)
    print(myCar.car_info())
    print(myCar.time_to_travel(100, 50))

```

### Multiple inheritance

```python
#child class inheriting more than one parent calss    

class Father:
    def __init__(self,name):
        self.father_name=name
    
    def father_info(self):
        return f"Father's name is {self.father_name}"

class Mother:
    def __init__(self,name):
        self.mother_name=name
    
    def mother_info(self):
        return f"Mother's name is {self.mother_name}"
    
class Child(Father,Mother):
    def __init__(self,father_name,mother_name,child_name):
        self.father_name=father_name
        self.mother_name=mother_name
        self.child_name=child_name
    
    def child_info(self):
        return f"{self.child_name} is the child of {self.father_name} and {self.mother_name}"
    
if __name__=="__main__":
    child=Child(father_name="John",mother_name="Jane",child_name="Jack")
    print(child.father_info())
    print(child.mother_info())
    print(child.child_info())
```

### Multilevel Inheritance

so the inherited class gets inherited by another class which creates a heirarchy

```python
from typing import List
class Directors:
    def __init__(self):
        self.directors={"director1":["manager1","manager2"]}
    
    def get_director(self,manager_name:str,managers:List)-> str:
        if manager_name in managers:
            for director, managers in self.directors.items():
                if manager_name in managers:
                    return f"{manager_name} is managed by {director}"
                else:
                    return f"{manager_name} is not managed by any director"
        else:
            return f"{manager_name} is not a manager in the system"
        

                
    def add_manager(self,director:str,manager_name:str):
        if director in self.directors:
            self.directors[director].append(manager_name)    
        else:
            self.directors[director]=[manager_name]

class Managers(Directors):

    def __init__(self):
        super().__init__()
        self.managers={"manager1":["John","Doe"],"manager2":["Jane","Smith"]}

    def get_manager(self,employee_name:str,employees:List)-> str:
        if employee_name in employees:
                for manager, employees in self.managers.items():
                    if employee_name in employees:
                        return f"{employee_name} is managed by {manager}"
                return f"{employee_name} is not managed by any manager"          
        else:
            return f"{employee_name} is not an employee in the system"
    
    def add_employee(self,manager:str,employee_name:str):
        if manager in self.managers:
            self.managers[manager].append(employee_name)    
        else:
            self.managers[manager]=[employee_name]
    
class employees(Managers):
    def __init__(self):
        super().__init__()
        self.employees={"John":"engineer","Doe":"designer","Jane":"analyst","Smith":"developer"}
    
    def get_designaion(self,employee_name:str)-> str:
        if employee_name in self.employees:
            return f"{employee_name} is a {self.employees[employee_name]}"
        else:
            return f"{employee_name} does not exist in the system"
    
    def add_employee(self, manager, employee_name, designation):
        super().add_employee(manager, employee_name)
        self.employees[employee_name] = designation


if __name__=="__main__":
    employee_system=employees()
    print(employee_system.get_designaion("John"))
    print(employee_system.get_manager("John",employee_system.employees.keys()))
    print(employee_system.get_director("manager1",employee_system.managers.keys()))
    print(employee_system.get_director("manager3",employee_system.managers.keys()))
    print(employee_system.get_designaion("Alice"))
```

### Compile time polymorphism

```python
class Calculator:
    def add(self,a: int|float,b:int|float,*args: int|float)-> int|float:
        total=a+b
        for num in args:
            total+=num
        return total
    
    def multiply(self,a: int|float,b:int|float,*args: int|float)-> int|float:
        total=a*b
        for num in args:
            total*=num
        return total

    def subtract(self,a: int|float,b:int|float,*args: int|float)-> int|float:
        total=a-b
        for num in args:
            total-=num
        return total

if __name__=="__main__":
    calc=Calculator()
    print(calc.add(1,2,3,4))
    print(calc.multiply(1,2,3,4))
    print(calc.subtract(10,1,2,3))
```

### Runtime overloading whhich is method overriding
same method name but different operations    

there's also method overloading

```python
class Animal:
    def sound(self):
        return "Some generic sound"

class Dog(Animal):
    def sound(self):
        return "Bark"

class Cat(Animal):
    def sound(self):
        return "Meow"

# Polymorphic behavior
dog=Dog()
print(dog.sound())  # Output: Bark
animals = [Dog(), Cat(), Animal()]
for animal in animals:
    print(animal.sound())
```

### Duck typing
if it looks like a duck and sounds like a duck it's a duck

so basically different class with same method name having similar return type are often considered as duck typing


```python
class Vehicle:
    def sound(self):
        return "honk honk"
    
class Animal:
    def sound(self):
        return "Some generic sound"

class Dog(Animal):
    def sound(self):
        return "Bark"

class Cat(Animal):
    def sound(self):
        return "Meow"
    
def make_sound(entity):
    print(entity.sound())


if __name__=="__main__":
    dog=Dog()
    cat=Cat()
    vehicle=Vehicle()

    make_sound(dog)      # Output: Bark
    make_sound(cat)      # Output: Meow
    make_sound(vehicle)  # Output: honk honk    
```

### Encapsulation
It's a way we define the accessibility of the data between public, private and protected

```python
class UserDetails:
    def __init__(self,name:str,password:str,email:str):
        self.name=name
        self.__password=password #private
        self._email=email #protected
    
    def get_password(self):
        return self.__password
    
    def set_password(self,password):
        self.__password=password


class User(UserDetails):
    def __init__(self,name:str,password:str,email:str):
        super().__init__(name,password,email)

    def get_email(self):
        return self._email

    def set_email(self,email):
        self._email=email

if __name__=="__main__": 
    user=User(name="Alice",password="password123",email="abc@gmail.com")
    print(user.name)  # Output: Alice
    print(user.get_password())  # Output: password123
    print(user.get_email())  # Output:
    user.set_password("newpassword456")
    print(user.get_password())  # Output: newpassword456
```

### Abstraction
it is about showing only what are essential to be shown and hiding the complex extra details or process

we use abstract class to implement this process

```python
from abc import ABC, abstractmethod
class Animal(ABC):
    
    @abstractmethod
    def sound(self):
        pass

    @property #abstract properties are for defining constant attributes we can't pass any value to it.
    @abstractmethod
    def legs(self):
        pass
    
    def info(self):#concrete method
        return f"This animal makes {self.sound()} sound and has {self.legs()} legs"

class Dog(Animal):
    def sound(self):
        return "Bark"
    
    def legs(self):
        return 4

class Monkey(Animal):
    def sound(self):
        return "Ooh Ooh Aah Aah"
    
    def legs(self):
        return 2
    

if __name__=="__main__":
    dog=Dog()
    monkey=Monkey()

    print(dog.info())  # Output: This animal makes Bark sound and has 4 legs
    print(monkey.info())  # Output: This animal makes Ooh Ooh Aah Aah sound and has 2 legs      
```
